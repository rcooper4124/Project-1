# What are the top 3 activities or personal experiences that caused students to take computing programs at CCM?

```python
df2024_selected = df2024.loc[:, "To what extent did the following activities or experience impact your decision to enroll in an computing course at CCM? [Middle/High school computing class]":"To what extent did the following activities or experience impact your decision to enroll in an computing course at CCM? [Other]"]
melted_df = df2024_selected.melt()

# Filter the melted DataFrame to only include "High Impact" responses
high_impact_counts = melted_df[melted_df['value'] == 'High Impact'].value_counts()


plt.figure(figsize=(12,6))  # Increase figure size for better visibility

# Plot bar chart for "High Impact" only
bars = high_impact_counts.plot(kind='bar', color=['blue'])

plt.xlabel("Impact Category")
plt.ylabel("Count")
plt.title("Distribution of Responses: How Students Became Interested in Computing Classes")

# Rotate x-ticks for better readability
plt.xticks(rotation=45, ha='right')

# Add legend
plt.legend(["High Impact"], title="Impact Levels", loc="upper right")

plt.show()
```

From this graph, we can see that the top 3 categories that influence a CCM student taking computing classes are, in order,
1. Family or friend influence
2. Family or friend working in the computing field
3. Middle/High school computing class
